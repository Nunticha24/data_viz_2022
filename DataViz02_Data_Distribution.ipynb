2. วาด boxplot เปรียบเทียบ การกระจายตัวของข้อมูล PetalLength ของดอก iris ทั้ง 3 ชนิด

###1. เขียน function ที่รับ input เป็น output ของ boxplot แล้ว แสดงค่า min, max, q1, q2, q3

```python
def box_vals(box_out_dict,vert=True) :
    if vert == True :   
        q1 = ybp['boxes'][0].get_ydata()[0]
        q2 = ybp['medians'][0].get_ydata()[0]
        q3 = ybp['boxes'][0].get_ydata()[2]
        iqr = q3 - q1
        mmin = min(ybp['caps'][0].get_ydata())
        mmax = max(ybp['caps'][1].get_ydata())
    else :
        q1 = ybp['boxes'][0].get_xdata()[0]
        q2 = ybp['medians'][0].get_xdata()[0]
        q3 = ybp['boxes'][0].get_xdata()[2]
        iqr = q3 - q1
        mmin = min(ybp['caps'][0].get_xdata())
        mmax = max(ybp['caps'][1].get_xdata())
    print(f'Q1 มีค่าเท่ากับ {q1}')
    print(f'Q2 มีค่าเท่ากับ {q2}')
    print(f'Q3 มีค่าเท่ากับ {q3}')
    print(f'Min มีค่าเท่ากับ {mmin}')
    print(f'Max มีค่าเท่ากับ {mmax}')
```

```python
box_vals(ybp,vert=False)
```
