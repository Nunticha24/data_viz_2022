#นำเข้าข้อมูล การตรวจโรค(ov)

```python
alldb =  pd.read_csv(os.path.join(path,'alldb.csv'))
alldb
```

```python
alldb.dtypes  
```

## เปลี่ยน types ของข้อมูล

###user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['user'].astype(str).str.isdigit()) & (alldb['user'].str.len() == 5)]
alldb
```

###RowKey เป็น String

```python
alldb['RowKey'] = alldb['RowKey'].astype('str')
alldb
```

###date เป็น YYYY-MM-DD

```python
import datetime #ทำการ import package datetime 
alldb = pd.DataFrame(alldb) 
alldb['date'] = pd.to_datetime(alldb['date'])
alldb
```

###phone เป็น String

```python
alldb['phone'] = alldb['phone'].astype('str')
alldb
```

###site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['site'].astype(str).str.isdigit()) & (alldb['site'].str.len() == 5)]
alldb
```

###time เป็น hh-mm-ss

```python
import datetime #ทำการ import package datetime 
alldb = pd.DataFrame(alldb) 
alldb['time'] = pd.to_datetime(alldb['time'],format= '%H:%M:%S' ).dt.time
alldb
```

###urine มีค่า String

```python
alldb['urine'] = alldb['urine'].astype('str')
alldb
```

###ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
alldb = alldb.dropna(subset=['ov'])
alldb
```

```python
set(alldb['ov'])
```

```python
alldb['ov'].replace({'ปกติ (เหลืองใส)':'0',
 'ขุ่น':'0',
 'ติดเชื้อ +2 ':'0',
 'ปกติ (เหลืองใส)':'0',
 'มีตะกอน':'0',
 'มีเม็ดเลือดแดง':'0',
 'สีน้ำนม':'0',
 'สีฟ้า หรือน้ำเงิน':'0',
 'ใสเป็นน้ำเปล่า':'0'}, inplace=True)
```

```python
set(alldb['ov'])
```

```python
alldb['ov'] = alldb['ov'].astype('int')
alldb
```

###PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb = alldb[(alldb['PartitionKey'].astype(str).str.isdigit()) & (alldb['PartitionKey'].str.len() == 5)]
alldb
```

```python
alldb_new= alldb[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey']] ## กำหนดตารางชื่อ alldb_df โดยเลือกเก็บข้อมูลในตาราง alldb จำนวน 9  คอลัมน์ 
alldb_new
```
