#นำเข้าข้อมูล การตรวจโรค(ov)

```python
alldb =  pd.read_csv(os.path.join(path,'alldb.csv'))
alldb
```

##เลือกเฉพาะ Column ที่ต้องการ

```python
alldb_new = alldb[['user','RowKey','date','phone','site','time','urine','ov','PartitionKey']]
alldb_new
```

###ใช้ .dropna ตัดแถวข้อมูลที่มีค่า missing

```python
alldb_drop =alldb_new.dropna()
alldb_drop
```

```python
alldb_drop.dtypes  
```

## เปลี่ยน type ของข้อมูล

###user เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb_drop = alldb_drop[(alldb_drop['user'].astype(str).str.isdigit()) & (alldb_drop['user'].str.len() == 5)]
alldb_drop
```

###RowKey เป็น String

```python
alldb_drop['RowKey'].astype({'RowKey':'string'}).dtypes
```

###date เป็น YYYY-MM-DD

```python
import datetime #ทำการ import package datetime 
alldb_drop = pd.DataFrame(alldb_drop)
alldb_drop['date'] = pd.to_datetime(alldb_drop['date'])
alldb_drop
```

###phone เป็น String

```python
alldb_drop['phone'].astype({'phone':'string'}).dtypes
```

###site เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb_drop = alldb_drop[(alldb_drop['site'].astype(str).str.isdigit()) & (alldb_drop['site'].str.len() == 5)]
alldb_drop
```

###time เป็น hh-mm-ss

```python
import datetime #ทำการ import package datetime 
alldb_drop = pd.DataFrame(alldb_drop) 
alldb_drop['time'] = pd.to_datetime(alldb_drop['time'],format= '%H:%M:%S' ).dt.time
alldb_drop
```

###urine มีค่า String

```python
alldb_drop['urine'].astype({'urine':'string'}).dtypes
```

###ov เป็นตัวเลข มีค่าตั้งแต่ 0-4 (0 คือไม่พบโรค 1-4 คือพบโรคที่ระดับความรุนแรงต่างๆ)

```python
alldb_drop = alldb_drop.dropna(subset=['ov'])
alldb_drop
```

```python
set(alldb_drop['ov'])
```

```python
alldb_drop['ov'].replace({'ติดเชื้อ +2 ':'0'}, inplace=True)
```

```python
set(alldb_drop['ov'])
```

```python
alldb_drop['ov'] = alldb_drop['ov'].astype('int')
alldb_drop
```

###PartitionKey เป็นตัวเลข 5 หลัก (ตัวเลขเท่านั้น)

```python
alldb_drop = alldb_drop[(alldb_drop['PartitionKey'].astype(str).str.isdigit()) & (alldb_drop['PartitionKey'].str.len() == 5)]
alldb_drop
```

```python
alldb_drop.dtypes
```

###แก้ Type

```python
alldb_drop['user'] = pd.to_numeric(alldb_drop['user'], errors='coerce')     
alldb_drop['date'] = pd.to_datetime(alldb_drop['date'],format="%d/%m/%Y")
alldb_drop['time'] = pd.to_datetime(alldb_drop['time'],format='%H:%M%:%S',errors='ignore')
alldb_drop['phone'] = alldb_drop['phone'].astype('string')
alldb_drop['urine'] = alldb_drop['urine'].astype('string')
alldb_drop['site'] = alldb_drop['site'].astype('int64')
alldb_drop['PartitionKey'] = alldb_drop['PartitionKey'].astype('int64')
alldb_drop['RowKey'] = alldb_drop['RowKey'].astype('string')
alldb_drop.dtypes
```

```python
alldb_drop
```

#นำเข้าข้อมูลโรงพยาบาล

```python
hospital_list =  pd.read_excel(os.path.join(path,'hospital_list.xlsx'))
hospital_list
```

```python
hospital_new = hospital_list[['hcode','name','zone_name','province','amphur','tambon','name5']]
hospital_new
```

##ตรวจสอบ Missing

```python
hospital_new.isnull().any()
```

```python
hospital_drop = hospital_new.dropna()
hospital_drop
```

```python
hospital_drop.isnull().sum()
```

```python
hospital_drop['name'] = hospital_drop['name'].fillna('ไม่ระบุ') 
hospital_drop['zone_name'] = hospital_drop['zone_name'].fillna('ไม่ระบุ') 
hospital_drop['amphur'] = hospital_drop['amphur'].fillna('ไม่ระบุ') 
hospital_drop['tambon'] = hospital_drop['tambon'].fillna('ไม่ระบุ') 
hospital_drop['name5'] = hospital_drop['name5'].fillna('ไม่ระบุ');
```

```python
hospital_drop.isnull().sum()
```

##เปลี่ยน hcode ให้เป็น int

```python
hospital_drop['hcode'] = pd.to_numeric(hospital_drop['hcode'], errors ='coerce')
```

```python
set(hospital_drop['hcode'])
```

```python
hospital_drop = hospital_drop.replace(np.nan, 0, regex=True)
```

```python
hospital_drop['hcode'] = hospital_drop['hcode'].astype('int')
```

```python
hospital_drop
```

##รวมตารางข้อมูลการตรวจโรค(ov) กับข้อมูลโรงพยาบาล


```python
alldata = alldb_drop.merge(hospital_drop,how='left',left_on='site',right_on='hcode')
alldata
```

```python
alldata.isnull().sum()
```

```python
alldata[alldata['hcode'].isnull()] 
```

```python
alldata = alldata.dropna()
alldata
```

```python
alldata.isnull().sum()
```

```python
alldata.to_csv('alldata.csv',index=False)
```
