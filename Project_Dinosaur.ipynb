###ตรวจสอบว่าข้อมูลซ้ำกันหรือไม่

าค่า min max ของ column NO.

```python
def min_max(i):
    a = min(i['No.'])
    b = max(i['No.'])
    print(f'min of NO. is {a}')
    print(f'max of NO. is {b}')
```

ดูค่า min max ของ covid ทั้ง 4 ตาราง

```python
min_max(data_covid2020)
```

```python
min_max(data_covid2021_1)
```

```python
min_max(data_covid2021_2)
```

```python
min_max(data_covid2022)
```

##รวมตาราง

ชื่อข้อมูลตารางที่พร้อมใช้งาน
* covid_dropped
* data_region
* Passenger



###ต่อตาราง covid ด้วยตาราง region

```python
covid_region = covid_dropped.merge(data_region,how='left',left_on='province_of_isolation',right_on='Province') 
print(covid_region.shape)
covid_region.head(3)
```

###ต่อตาราง covid ด้วยตาราง region

```python
all_data = covid_region.merge(Passenger,how='left',left_on='announce_date',right_on='วันที่')     # ต่อตาราง covid ด้วยตาราง region
print(all_data.shape)
all_data.head(3)
```

```python
all_data.isnull().any()
```

#เปลี่ยนชนิดของข้อมูล

```python
all_data['announce_date'] = pd.to_datetime(all_data['announce_date'],format="%d/%m/%Y")
```

```python
all_data.dtypes
```

```python
all_data
```

#แยกข้อมูล
เนื่องจากขนาดไฟล์ใหญ่เกินไปจึงแยกข้อมูลออกเป็นส่วนๆ

```python
covid2020 = all_data[all_data['announce_date'] < '2021-01-01']  #เลือกข้อมูลตั้งแต่วันแรก ถึง วันที่ 2020-12-31
print(covid2020.shape)                                            
covid2020.head(3)
```

```python
covid2021_1 = all_data[(all_data['announce_date'] >= '2021-01-01') & (all_data['announce_date'] < '2021-04-01')] #เลือกข้อมูลไตรมาสแรกของปี 2021
print(covid2021_1.shape)
covid2021_1.head(3)
```

```python
covid2021_2 = all_data[(all_data['announce_date'] >= '2021-04-01') & (all_data['announce_date'] < '2021-07-01')] #เลือกข้อมูลไตรมาสสองของปี 2021
print(covid2021_2.shape)
covid2021_2.head(3)
```

```python
covid2021_3 = all_data[(all_data['announce_date'] >= '2021-07-01') & (all_data['announce_date'] < '2021-10-01')] #เลือกข้อมูลไตรมาสสามของปี 2021
print(covid2021_3.shape)
covid2021_3.head(3)
```

```python
covid2021_4 = all_data[(all_data['announce_date'] >= '2021-10-01') & (all_data['announce_date'] < '2022-01-01')] #เลือกข้อมูลไตรมาสสี่ของปี 2021
print(covid2021_4.shape)
covid2021_4.head(3)
```

```python
covid2022_1 = all_data[(all_data['announce_date'] >= '2022-01-01') & (all_data['announce_date'] < '2022-02-01')] #เลือกข้อมูลเดือนมกราคม 2022
print(covid2022_1.shape)
covid2022_1.head(3)
```

```python
covid2022_2 = all_data[(all_data['announce_date'] >= '2022-02-01') & (all_data['announce_date'] < '2022-03-01')] #เลือกข้อมูลเดือนกุมภาพันธ์ 2022
print(covid2022_2.shape)
covid2022_2.head(3)
```

```python
covid2022_3 = all_data[all_data['announce_date'] >= '2022-03-01'] #เลือกข้อมูลเดือนมีนาคม 2022 ถึงล่าสุดที่โหลดมา
print(covid2022_3.shape)
covid2022_3.head(3)
```
