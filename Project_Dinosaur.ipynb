#นำเข้าข้อมูลที่เลือกเอง

```python
passenger =  pd.read_csv(os.path.join(path,'passenger-covid19.csv'))
passenger
```

###เลือกเฉพาะ column ที่ต้องการ

```python
Passenger = passenger[['วันที่','ทางถนน (คน)','ทางราง (คน)','ทางน้ำ (คน)', 'ทางอากาศ* (คน)', 'รวมประชาชนเดินทาง (คน)']]
Passenger
```

##รวมตาราง covid_region กับตาราง Passenger


```python
all_data = covid_region.merge(Passenger,how='left',left_on='announce_date',right_on='วันที่')
all_data
```

##ตรวจสอบค่า Missing

```python
all_data.isnull().any() #ดูค่า missing ของแต่ละ column
```

### เปลี่ยน datetime เพื่อที่จะแบ่งข้อมูลออกไปใช้งาน

```python
import datetime
all_data = pd.DataFrame(all_data)
all_data['announce_date'] = pd.to_datetime(all_data['announce_date'], format = "%d/%m/%Y")
all_data
```

```python
all_data.dtypes
```

```python
all_data
```

##แบ่งข้อมูลไปใช้งาน

```python
covid2020 = all_data[(all_data['announce_date'] >= '2020-01-12') & (all_data['announce_date'] <= '2020-12-31')] #ข้อมูลปี 2020
covid2020
```

```python
covid2021_1 = all_data[(all_data['announce_date'] >= '2021-01-01') & (all_data['announce_date'] <= '2021-03-31')] #ข้อมูลไตรมาสแรกของปี 2021
covid2021_1
```

```python
covid2021_2 = all_data[(all_data['announce_date'] >= '2021-04-01') & (all_data['announce_date'] <= '2021-06-30')] #ข้อมูลไตรมาสสองของปี 2021
covid2021_2
```

```python
covid2021_3 = all_data[(all_data['announce_date'] >= '2021-07-01') & (all_data['announce_date'] <= '2021-09-30')] #ข้อมูลไตรมาสสามของปี 2021
covid2021_3
```

```python
covid2021_4 = all_data[(all_data['announce_date'] >= '2021-10-01') & (all_data['announce_date'] <= '2021-12-31')] #ข้อมูลไตรมาสสี่ของปี 2021
covid2021_4
```

```python
covid2022_1 = all_data[(all_data['announce_date'] >= '2022-01-01') & (all_data['announce_date'] <= '2022-01-31')] #ข้อมูลเดือนมกราคมปี 2022
covid2022_1
```

```python
covid2022_2 = all_data[(all_data['announce_date'] >= '2022-02-01') & (all_data['announce_date'] <= '2022-02-28')] #ข้อมูลเดือนกุมภาพันธ์ปี 2022
covid2022_2
```

```python
covid2022_3 = all_data[all_data['announce_date'] >= '2022-03-01'] #ข้อมูลวันที่ 1- 18 เดือนมีนาคม ปี 2022
covid2022_3
```

##เลือกทำข้อมูลวันที่ 1-18 มีนาคม 2022
ซึ่งเป็นข้อมูลล่าสุด

###ทำการ Groupby เพื่อดูจำนวนผู้รักษาตัวสะสมของแต่ละจังหวัด ตั้งแต่วันที่ 1-18 มีนาคม 2022

```python
data_gb_2022_3 = covid2022_3.groupby('province_of_isolation').size().reset_index(name='จำนวนผู้รักษาตัว')
data_gb_2022_3
```

```python
Covid2022_3 = covid2022_3.merge(data_gb_2022_3,how='left',left_on='province_of_isolation',right_on = 'province_of_isolation') 
Covid2022_3
```
